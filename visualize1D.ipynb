```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
funcs = [
    lambda x: np.exp(np.sin(x))
]
```

```python
with open('./Multigrid/result.txt') as f:
    # our solution
    vals = np.array([float(s) for s in f.readlines()])
    n = len(vals) - 1
    
    # ground truth
    x = np.arange(0.0, 1+0.5/n, 1/n)
    ground_truth = funcs[0](x) - (funcs[0](0) * (1 - x) + funcs[0](1) * x)
    
    # show
    plt.figure(dpi=150)
    plt.plot(x, vals, label='Our Solution')
    plt.plot(x, ground_truth, label='Ground Truth')
    plt.title('Result')
    plt.legend()
    plt.show()

    delta = vals - ground_truth
    print('max-norm:', delta.max())
```
