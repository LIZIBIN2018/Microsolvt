```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
funcs = [
    lambda x: np.exp(np.sin(x))
]
```

```python
for times in range(5, 9):
    filepath = "../Multigrid/result1D_" + str(2**times) + ".txt"
    print(filepath)

    with open('../Multigrid/result1D_32.txt') as f:
        # our solution
        for l in f.readlines():
            ll = l.split(" ")

            aim = []
            for i in ll:
                aim.append((float)(i))
        vals = np.array(aim)
        n = len(vals) - 1

        # ground truth
        x = np.arange(0.0, 1+0.5/n, 1/n)
        ground_truth = funcs[0](x) - (funcs[0](0) * (1 - x) + funcs[0](1) * x)
        
        # show
        plt.figure(dpi=150)
        plt.plot(x, vals, label='Our Solution')
        plt.plot(x, ground_truth, label='Ground Truth')
        plt.title('Result')
        plt.legend()
        plt.show()

        delta = vals - ground_truth
        print('max-norm:', delta.max())
        
```
